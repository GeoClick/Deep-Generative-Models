```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import librosa
```

# Data : Sine Wave

```python
sine_x = np.linspace(0, 100000, 10000000)
sine_wave = np.sin(sine_x)

def get_data(batch, time_length):
    start_index = np.random.randint(0, len(sine_wave) - (batch * time_length), 1)[0]
    end_index = start_index + (batch * time_length)
    data = sine_wave[start_index:end_index]
    data = np.reshape(data, [batch, time_length])
    return data

data = get_data(10, 8000)
print(data.shape)

plt.figure(figsize=[18, 3])
plt.plot(data[0])
plt.plot(data[1])
plt.show()
```

# Hyper-parameters

```python
Batch = None
TimeLength = None

# teacher wavenet params
Teacher_Filter_Width = 2
Teacher_Dilations = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512,
                    1, 2, 4, 8, 16, 32, 64, 128, 256, 512]

# student wavenet params
Student_Filter_Width = 3
Student_Dilations = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]
Student_IAF_Layers = 4

# wavenet common params
Hidden_channels = 8
Skip_channels = 16
Out_channels = 2
```

# Naive Wavenet

```python
def conv(x, filters, kernel_size=1, dilation=1, stddev=0.2):
    return tf.layers.conv1d(x, filters=filters, kernel_size=kernel_size, dilation_rate=dilation,
                                  kernel_initializer=tf.initializers.truncated_normal(stddev=stddev))
    
def wavenet(x, filter_width, dilations, 
            hidden_channels, skip_channels, out_channels, 
            name='wavenet', reuse=False):
    
    with tf.variable_scope(name, reuse=reuse):

        outputs = conv(x, skip_channels)
        
        skips = []
        
        for i, dilation in enumerate(dilations):
            layer_inputs_paded = tf.pad(outputs, [[0, 0], [(filter_width - 1) * dilation, 0], [0, 0]])
            
            filter_outputs = conv(layer_inputs_paded, hidden_channels, filter_width, dilation)
            gate_outputs = conv(layer_inputs_paded, hidden_channels, filter_width, dilation)
            
            layer_outputs = tf.nn.tanh(filter_outputs) * tf.nn.sigmoid(gate_outputs)
            
            outputs += conv(layer_outputs, skip_channels)
            skips.append(conv(layer_outputs, skip_channels))
            
        outputs = sum(skips)
        outputs = tf.nn.relu(outputs)
        outputs = conv(outputs, skip_channels)
        outputs = tf.nn.relu(outputs)
        outputs = conv(outputs, out_channels)
        
        return outputs
    
    
def student_wavenet(z):
    mean_total = tf.zeros_like(z)
    log_scale_total = tf.zeros_like(z)
    
    z_list = [z]
    
    for i in range(Student_IAF_Layers):
        z_less = tf.pad(z[:, :-1], [[0, 0], [1, 0], [0, 0]])
        y = wavenet(z_less, Student_Filter_Width, Student_Dilations, 
                Hidden_channels, Skip_channels, Out_channels, name='StudentWavenet' + str(i))
        mean, log_scale = y[:, :, 0:1], y[:, :, 1:2]
        log_scale = tf.maximum(log_scale, -8.)
        scale = tf.exp(log_scale)
        
        z = z * scale + mean
        z_list.append(z)
        mean_total = mean_total * scale + mean
        log_scale_total += log_scale
        
    x = z
    return x, z_list, mean_total, log_scale_total
    
```

# Draw Graph

```python
tf.reset_default_graph()

'''
Input Layer
'''

x = tf.placeholder(dtype=tf.float32, shape=[Batch, TimeLength], name='x')

'''
Teacher Wavenet Layer
'''

y = wavenet(x[:, :, None], Teacher_Filter_Width, Teacher_Dilations, 
            Hidden_channels, Skip_channels, Out_channels, name='TeacherWavenet')
mean, log_scale = y[:, :-1, 0], y[:, :-1, 1]
log_scale = tf.maximum(log_scale, -8.)
scale = tf.exp(log_scale)
var = scale ** 2

'''
Teacher Loss Layer
'''
teacher_dist = tf.distributions.Normal(loc=mean, scale=scale)
nll = -teacher_dist.log_prob(x[:, 1:])
nll = tf.reduce_mean(nll)
sample = teacher_dist.sample()

teacher_global_step = tf.Variable(0, name='teacher_global_step')
teacher_learning_rate = tf.placeholder(dtype=tf.float32, shape=())
teacher_train_step = tf.train.AdamOptimizer(teacher_learning_rate).minimize(nll, teacher_global_step)

'''
Student Wavenet Layer
'''
Student_Batch = 1
Student_TimeLength = 8000

z = tf.random.normal([Student_Batch, Student_TimeLength, 1])
student_x, z_list, student_mean, student_log_scale = student_wavenet(z)
student_log_scale = tf.maximum(student_log_scale, -8.)
student_scale = tf.exp(student_log_scale)
student_dist = tf.distributions.Normal(loc=student_mean[:, 1:], scale=student_scale[:, 1:])

### Put student's samples into teacher wavenet

y = wavenet(student_x, Teacher_Filter_Width, Teacher_Dilations, 
            Hidden_channels, Skip_channels, Out_channels, name='TeacherWavenet', reuse=True)
y = tf.stop_gradient(y)

teacher_mean, teacher_log_scale = y[:, :-1, 0:1], y[:, :-1, 1:2]
teacher_log_scale = tf.maximum(teacher_log_scale, -8.)
teacher_scale = tf.exp(teacher_log_scale)
teacher_dist = tf.distributions.Normal(loc=teacher_mean, scale=teacher_scale)

'''
Student Loss Layer
'''
student_log_scale = student_log_scale[:, 1:]
student_scale = student_scale[:, 1:]
student_mean = student_mean[:, 1:]

kl_divergence = tf.distributions.kl_divergence(student_dist, teacher_dist)
kl_divergence = tf.reduce_mean(kl_divergence)

student_global_step = tf.Variable(0, name='student_global_step')
student_learning_rate = tf.placeholder(dtype=tf.float32, shape=())
student_train_step = tf.train.AdamOptimizer(student_learning_rate).minimize(kl_divergence, student_global_step)



'''
Session Open
'''

sess = tf.Session()
sess.run(tf.global_variables_initializer())

print('graph created')
```

# Teacher & Student Training Loop

```python
from IPython.display import clear_output
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt

# training parameters

loops = 100
batch = 2
time_length = 8000
_learning_rate = 1e-4

while(True):
    _nlls = []
    _kls = []
    for _ in tqdm(range(loops)):
        _x = get_data(batch=batch, time_length=time_length)
        fetches = [teacher_train_step, teacher_global_step, nll, sample]
        _, _teacher_global_step, _nll, _sample = sess.run(fetches, 
                                                         feed_dict={x: _x,
                                                                    teacher_learning_rate: _learning_rate})
        _nlls.append(_nll)
        
        fetches = [student_train_step, student_global_step, kl_divergence, z_list]
        _, _student_global_step, _kl, _z_list = sess.run(fetches, 
                                                            feed_dict={student_learning_rate: 1e-4})
        
        _kls.append(_kl)
                
    clear_output()
    
    print('teacher step :', _teacher_global_step, 'negative log-likelihood :', np.mean(_nlls))
    print('student step :', _student_global_step, 'kl-divergence :', np.mean(_kls))
    
    print('true wave & teacher\'s prediction')
    plt.figure(figsize=[18, 2])
    plt.plot(_x[0], alpha=0.5)
    plt.plot(_sample[0], alpha=0.5)
    plt.show()
    
    print('student\'s outputs')
    for i, _z in enumerate(reversed(_z_list)):
        print('z' + str(len(_z_list) - i - 1))
        plt.figure(figsize=[18, 2])
        plt.plot(_z[0, :, 0], alpha=0.5)
        plt.show()
    
    if _student_global_step > 1000:
        break
        
```
